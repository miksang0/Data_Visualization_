```python
import seaborn as sns
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
tips
```

```python
sns.scatterplot(x="tip", y="total_bill", data=tips, hue = "day", size = "size", palette="YlGnBu")
```

```python
sns.histplot(data=tips, x="tip", kde=True, bins = 15)
```

```python
sns.displot(data=tips, x="tip", kde=True, bins = 15)
```

```python
sns.barplot(data = tips, x = 'day', y = 'tip')
```

```python
sns.boxplot(data = tips, x = "day", y = "total_bill", hue = "sex", palette="YlGnBu")
```

```python
sns.stripplot(data = tips, x = "day", y = "total_bill", hue = "sex", dodge = True, palette="YlGnBu")
```

```python
#kind could be anything here. (reg for regression, hex for hexagons, kde for kernal density estimate)
#No fill in hex. Fill shows the density.
sns.jointplot(data = tips, x = "tip", y = "total_bill", kind = "hex", cmap = "YlGnBu")
```

```python
sns.pairplot(titanic.select_dtypes(['number']), hue = "pclass")
```

```python
titanic.corr(numeric_only = True)
```

```python
sns.heatmap(titanic.corr(numeric_only = True), annot=True, cmap = "coolwarm")
```
