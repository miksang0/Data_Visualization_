```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/StudentsPerformance.csv')
df.head()
```

```python
df.groupby('race/ethnicity')['math score'].mean().plot()
```

```python
df_race = df.groupby('race/ethnicity').agg({'math score': ['mean']})
df_race
```

```python
df_race.columns = ['Average Math Score']
df_race
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(df_race.index, df_race['Average Math Score'])
plt.xlabel('Race/Ethnicity')
plt.ylabel('Average Math Score')
plt.title('Average Math Score by Race/Ethnicity')
plt.show()
```
