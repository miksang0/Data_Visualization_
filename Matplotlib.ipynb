```python
import numpy as np
import matplotlib.pyplot as plt

X_data = np.random.random(100) * 100
y_data = np.random.random(100) * 100

plt.scatter(X_data, y_data, c ="green", s = 120, alpha = 0.3)
plt.show()
```

```python
years = [2003 + x for x in range(16) ]
weight = [85, 89, 85, 87, 88, 89, 95, 98, 95, 81, 80, 85, 99, 101, 95, 91]
plt.plot(years, weight, c = 'g', lw = 2, linestyle = '--')
plt.show()
```

```python
x = ["C++", "Python", "Java", "Go", "C#" ]
y = [50, 80, 60, 20, 25]
plt.bar(x, y, color = 'green', align = 'edge', width = 0.5, edgecolor = 'red', lw = 5)
plt.show()
```

```python
ages = np.random.normal(20, 1.5, 1000)
plt.hist(ages, bins = 15, cumulative = True)
plt.show()
```

```python
langs = ["C++", "Python", "Java", "Go", "C#" ]
votes = [75, 90, 20, 50, 45]
explodes = [0, 0, 0.2, 0, 0]
plt.pie(votes, labels = langs, explode = explodes, autopct = "%.2f%%", startangle = 90)
plt.show()
```
